```python
import os
import pandas as pd
import glob
import json
version = "v4"
data_path = f"../data/interim/{version}"
dataset_target_path = "../data/processed"
```

```python
from lib import read_dataset
data_on_raw, data_off_raw = read_dataset(data_path, classes = 2)
```

```python
from lib import add_intensity_to_dataset
add_intensity_to_dataset(data_on_raw)
add_intensity_to_dataset(data_off_raw)
```

```python
from collections import ChainMap
def subsamples(df: pd.DataFrame, name: str, window=2000, stride=10):
    return {f"{name}_{i:03d}": df[start:start+window] for i, start in enumerate(range(0,df.shape[0]-window, stride))}

data_on = dict(ChainMap(*[subsamples(df, name) for name,df in data_on_raw.items()]))
data_off = dict(ChainMap(*[subsamples(df, name) for name,df in data_off_raw.items()]))
```

```python
import time
from lib import save_dataset
with open(f"{data_path}/metadata.json", 'r') as jsonfile:
    metadata = json.load(jsonfile)
    metadata['title'] = f"{version} fourier dataset big"
    metadata['description'] = 'Fourier dataset with windowed test cases from '+metadata['description']
    metadata['format'] = "processed"
    metadata['author'] = 'Mihael'
    metadata['version'] = version
    metadata['created'] = int(time.time())
save_dataset(data_fourier, dataset_target_path, dataset_name = 'fourier_window_big', metadata=metadata)
```
