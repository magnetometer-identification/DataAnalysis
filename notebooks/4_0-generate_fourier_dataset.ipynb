```python
from collections import ChainMap
def subsamples(df: pd.DataFrame, name: str, window=2000, stride=10):
    return {f"{name}_{i:03d}": df[start:start+window] for i, start in enumerate(range(0,df.shape[0]-window, stride))}

data_on = dict(ChainMap(*[subsamples(df, name) for name,df in data_on_raw.items()]))
data_off = dict(ChainMap(*[subsamples(df, name) for name,df in data_off_raw.items()]))
```

```python
from lib import spectral_dataset
dataset_on_intensity = spectral_dataset(data_on, 6, vec_len=50, column='Intensity', experiment_duration=10)
dataset_on_X = spectral_dataset(data_on, 6, vec_len=50, column='X_UnCal', experiment_duration=10)
dataset_on_Y = spectral_dataset(data_on, 6, vec_len=50, column='Y_UnCal', experiment_duration=10)
dataset_on_Z = spectral_dataset(data_on, 6, vec_len=50, column='Z_UnCal', experiment_duration=10)
dataset_off_intensity = spectral_dataset(data_off, 5, vec_len=50, column='Intensity', experiment_duration=10)
dataset_off_X = spectral_dataset(data_off, 5, vec_len=50, column='X_UnCal', experiment_duration=10)
dataset_off_Y = spectral_dataset(data_off, 5, vec_len=50, column='Y_UnCal', experiment_duration=10)
dataset_off_Z = spectral_dataset(data_off, 5, vec_len=50, column='Z_UnCal', experiment_duration=10)
```

```python
dataset_on_intensity.rename(lambda x: f"intensity_{x}", axis='columns', inplace=True)
dataset_on_X.rename(lambda x: f"X_{x}", axis='columns', inplace=True)
dataset_on_Y.rename(lambda x: f"Y_{x}", axis='columns', inplace=True)
dataset_on_Z.rename(lambda x: f"Z_{x}", axis='columns', inplace=True)
```

```python
dataset_off_intensity.rename(lambda x: f"intensity_{x}", axis='columns', inplace=True)
dataset_off_X.rename(lambda x: f"X_{x}", axis='columns', inplace=True)
dataset_off_Y.rename(lambda x: f"Y_{x}", axis='columns', inplace=True)
dataset_off_Z.rename(lambda x: f"Z_{x}", axis='columns', inplace=True)
```

```python
dataset_on = pd.concat([dataset_on_intensity, dataset_on_X.iloc[:,4:], dataset_on_Y.iloc[:,4:], dataset_on_Z.iloc[:,4:]], axis=1)
dataset_on = dataset_on.rename(columns={"intensity_name": "name", "intensity_location": "location", "intensity_device_id": "device_id", "intensity_label": "label"})
```

```python
dataset_off = pd.concat([dataset_off_intensity, dataset_off_X.iloc[:,4:], dataset_off_Y.iloc[:,4:], dataset_off_Z.iloc[:,4:]], axis=1)
dataset_off = dataset_off.rename(columns={"intensity_name": "name", "intensity_location": "location", "intensity_device_id": "device_id", "intensity_label": "label"})
```
